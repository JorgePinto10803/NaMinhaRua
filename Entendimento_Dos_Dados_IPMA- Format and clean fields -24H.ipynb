```python
#///Combine de 4 excel files into one data set
base_dir = Path.home()
ipma_source_directory_path = base_dir / "NaMinhaRua" / "Ficheiros de dados originais" / "IPMA"
# Change the current working directory to the specified directory
os.chdir(ipma_source_directory_path)
# Get a list of all Excel files in the directory
excel_files = glob.glob('*.csv')
# Print the list of Excel files
dfs = []
# Iterate through each Excel file and read it into a DataFrame
for file in excel_files:
    df = pd.read_csv(file)
    dfs.append(df)
# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)
# Write the combined DataFrame to a single Excel file or CSV file
manipulated_data_directory = base_dir / "NaMinhaRua" / "Ficheiros de dados Manipulados"
# Write the result
combined_df.to_csv(manipulated_data_directory / 'combined_ipma_data.csv',sep=',', index=False)
```

```python
pd.options.display.max_rows = 9999999999
#////////////////////////////////////////////////////////Import Dataset/////////////////////////////////////////////////////////////
base_dir = Path.home()
manipulated_data_directory = base_dir / "NaMinhaRua" / "Ficheiros de dados Manipulados"
csv_path=manipulated_data_directory
ipmadataset= pd.read_csv(csv_path / 'combined_ipma_data.csv')
ipmadataset = pd.DataFrame(ipmadataset)
print ("\nMissing values :  ", ipmadataset.isnull().sum().values.sum())
```

```python
# Remove "entity_ts" column which represents the data extraction time, with a gap of several hours comparing with "fecha"
ipmadataset.head(12)
ipmadataset = ipmadataset.drop(columns=['entity_id', 'entity_ts'])
ipmadataset.head(12)



```

```python
print(ipmadataset['estacion'].unique())

ipmadataset['estacion'] = ipmadataset['estacion'].astype(str)
mapeamento = {'1200579': '579', '1200535': '535', '1210762': '762'}

# Replace estacion column values with the correct mapping for the 3 estacions
ipmadataset['estacion'] = ipmadataset['estacion'].replace(mapeamento)
# Show values after the replacement
print("Valores únicos na coluna 'estacion' após a substituição:")
print(ipmadataset['estacion'].unique())
```

```python
# Convert "fecha" to datetime format
ipmadataset['fecha'] = pd.to_datetime(ipmadataset['fecha'])

# Filter and exclude the error values
ipmadataset = ipmadataset[ipmadataset['precacumulada'] != -99.0]

# Add a column with date( without hour) to facilitate the daily agregation
ipmadataset['data_precacumulada_media_diaria'] = ipmadataset['fecha'].dt.date

# Group by estacion and date, adding the 'precacumulated' values
soma_precacumulada_diaria = ipmadataset.groupby(['estacion', 'data_precacumulada_media_diaria'])['precacumulada'].sum().reset_index()

# Create the new variable by dividing the daily sum by 24 and renaming the column
soma_precacumulada_diaria['media_precacumulada_horaria_diaria'] = soma_precacumulada_diaria['precacumulada'] / 24

# Print the result
print(soma_precacumulada_diaria)

# add new column "media_precacumulada_horaria_diaria" to the dataset
ipmadataset = pd.merge(ipmadataset, 
                       soma_precacumulada_diaria[['estacion', 'data_precacumulada_media_diaria', 'media_precacumulada_horaria_diaria']], 
                       on=['estacion', 'data_precacumulada_media_diaria'], 
                       how='left')

print("///////////////////////////////////////////////////////////////////////////////////////////")
# Result
print(ipmadataset[['estacion', 'data_precacumulada_media_diaria', 'media_precacumulada_horaria_diaria']])
```

```python
# Determine the highest average accumulated precipitation by season and the day it occurred
max_media_precacumulada_por_estacao = soma_precacumulada_diaria.loc[
    soma_precacumulada_diaria.groupby('estacion')['media_precacumulada_horaria_diaria'].idxmax()
].reset_index(drop=True)

# Result
print("///////////////////////////////////////////////////////////////////////////////////////////")
print(max_media_precacumulada_por_estacao)
```
